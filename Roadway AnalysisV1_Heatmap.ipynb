```python
import pandas as pd

friction_df = pd.read_excel('friction_data.xlsx')
accidents_df = pd.read_excel('accidents.xlsx')
```

```python
friction_df['date'] = pd.to_datetime(friction_df['date'])

friction_df['year'] = friction_df['date'].dt.year

friction_df.info()
```

```python
accidents_df['Collision Date'] = pd.to_datetime(accidents_df['Collision Date'])

accidents_df['year'] = accidents_df['Collision Date'].dt.year

accidents_df.info()
```

```python
import folium

EECP_map = folium.Map(location=[38.339, -85.674], zoom_start=11)
```

```python
EECP_map
```

```python
fric_max = friction_df['friction_number'].max()
fric_min = friction_df['friction_number'].min()

friction_df['weight'] = (fric_max - friction_df['friction_number']) / (fric_max - fric_min)
```

```python
fric_min
```

```python
fric_max
```

```python
years = sorted(friction_df['year'].unique())
```

```python
heat_data = []      
time = []     
for yr in years:
    yearly_points = friction_df[friction_df['year'] == yr][['latitude', 'longitude', 'weight']].values.tolist()
    heat_data.append(yearly_points)
    time.append(str(yr))
```

```python
time
```

```python
from folium import plugins
```

```python
plugins.HeatMapWithTime( heat_data, index=time, radius=60, max_opacity=0.8, name="Making Connections: Accidents vs Pavement Roughness").add_to(EECP_map)
EECP_map
```
